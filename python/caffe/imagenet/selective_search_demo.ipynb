First of all, we'll need a little [Python script](https://github.com/sergeyk/selective_search_ijcv_with_python) to run the Matlab Selective Search code.

Let's run detection on an image of a cat lounging on a couch (one of the ImageNet pictures), which we will download from the web.

    wget http://farm2.static.flickr.com/1104/1204206441_bc256e34f2.jpg
    echo "1204206441_bc256e34f2.jpg" > image_cat.txt
    python power_wrapper.py --images_file=image_cat.txt --crop_mode=selective_search --model_def=../../../examples/imagenet_deploy.prototxt --pretrained_model=alexnet_train_iter_470000 --output=selective_cat.h5
    
Now we can load the results to visualize them.
You'll see we store filenames, windows, and computed ImageNet predictions.
Of course, we only ran on one image, so the filenames are all the same.
In general, `power_wrapper` is most efficient when running on a lot of images: it first extracts window proposals for all of them, then batches the windows for efficient GPU processing, and then outputs the results.
